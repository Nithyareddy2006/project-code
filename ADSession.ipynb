```python
import pandas as pd
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

```

```python
## s and y
x=df.iloc[:,:8]
y=df.iloc[:,8]
```

```python
x
```

```python
y
```

```python
seed=7
test_sze=0.30
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=test_sze,random_state=seed)
```

```python
seed
```

```python
##train the model
model=XGBClassifier()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
predictions=[round(value) for value in y_pred]
print(predictions)
```

```python
acciracy=accuracy_score(y_test,predictions)
print("Accuracy:%.2f%%"%(acciracy*100.0))
```
